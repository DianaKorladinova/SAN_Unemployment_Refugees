### Process Salaries till 2010

```python
quarters = {
    "Q1": [1, 2, 3],
    "Q2": [4, 5, 6],
    "Q3": [7, 8, 9],
    "Q4": [10, 11, 12],
}
region_codes_map = {
    '3018': 'PHA',
    '3026': 'STC',
    '3034': 'JHC',
    '3042': 'PLK',
    '3051': 'KVK',
    '3115': 'JHM',
    '3107': 'VYS',
    '3085': 'HKK',
    '3077': 'LBK',
    '3140': 'MSK',
    '3123': 'OLK',
    '3093': 'PAK',
    '3069': 'ULK',
    '3131': 'ZLK'
}
region_code = ['3018', '3034', '3115', '3051', '3107', '3085', '3077', '3140', '3123', '3093', '3042', '3026', '3069',
               '3131']
```

```python
def get_salaries(region_code):
    sal_df = pd.DataFrame(columns=('month', 'year', 'kraj', 'avg_monthly_salary'))
    count = 0
    for i in range(0, len(region_code)):
        path = f"https://vdb.czso.cz/vdbvo2/faces/cs/index.jsf?page=vystup-objekt&z=T&f=TABULKA&skupId=855&filtr=G%7EF_M%7EF_Z%7EF_R%7ET_P%7E_S%7E_null_null_&katalog=30852&pvo=MZD01-E&pvo=MZD01-E&evo=v208_!_MZD01-Q-do2010_1&u=v479__VUZEMI__100__{region_code[i]}#w="
        #path = f"https://vdb.czso.cz/vdbvo2/faces/cs/index.jsf?page=vystup-objekt&pvo=MZD01-D&z=T&f=TABULKA&skupId=855&filtr=G~F_M~F_Z~F_R~T_P~_S~_null_null_&katalog=30852&u=v720__VUZEMI__100__{region_code[i]}&&evo=v750_!_MZD-Q-INDX-od2011_1&str=v720"
        try:
            df = pd.read_html(path, decimal=",", thousands="\u00A0")[6][["Unnamed: 0_level_0", "Průměrná měsíční mzda (na přepočtené počty zaměstnanců)"]].droplevel(level=1, axis=1)
            # Iterate over rows using iterrows
            for index, row in df.iterrows():
                time_data = str.split(row['Unnamed: 0_level_0'])
                year = time_data[0]
                # get the first two chars
                quarter = time_data[1][:2]
                qrtr_months = quarters.get(quarter)
                for month in qrtr_months:
                    dict = {
                        'month': int(month),
                        'year': int(year),
                        'kraj': region_codes_map.get(region_code[i]),
                        'avg_monthly_salary': df["Průměrná měsíční mzda (na přepočtené počty zaměstnanců)"].iloc[index]
                    }
                    sal_df.loc[count] = dict
                    count += 1
        except Exception as e:
            print("error:", e, " ", " ", region_code[i])
    return sal_df
```

```python
sal_df = get_salaries(region_code)
sal_df
```

```python
sal_df2 = sal_df.loc[(sal_df["year"] >= 2008)]
sal_df2.sort_values(by=["year", "month", "kraj"], inplace=True)
sal_df2.set_index(keys=["month", "year", "kraj"], inplace=True)
sal_df2
```

```python
crim_sal_df = pd.concat([df2, sal_df2], axis=0)
crim_sal_df = crim_sal_df.groupby(by=["month", "year", "kraj"]).sum()
#crim_sal_df = crim_sal_df.map(lambda x: np.nan if x == 0.0 else x)
crim_sal_df
```

## Merge All Together

```python
import os
res_df = None
#path = "../data/prepared"
path = "../../dataset/sub_datasets/"
for file in os.listdir(path):
    tmp_df = pd.read_csv(os.path.join(path, file), index_col=["month", "year", "kraj"])
    display(tmp_df.head())
    if res_df is None:
        res_df = tmp_df
    else:
        res_df = pd.concat([res_df, tmp_df], axis=1)

display(res_df)
```
