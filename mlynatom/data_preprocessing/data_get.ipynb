## Refugees - MV

```python
def prepare_month(month: str, year: str, prev_df: Optional[pd.DataFrame]=None) -> pd.DataFrame:
    for day in range(month2days[month], 0, -1):
        for path in [f"https://www.mvcr.cz/soubor/statistika-ukr-{day}-{month}-{year}-xls.aspx", f"https://www.mvcr.cz/soubor/statistika-obce-suk-vek-vse-{day}-{month}-{year}-xls.aspx", f"https://www.mvcr.cz/soubor/statistika-ukr-{day}-{month}-{year}-xlt.aspx"]:
            try:
                df = pd.read_excel(path)
                df.drop(columns=["okres", "obec", "kod_obce", "obec_mc", "m_do_3", "z_do_3", "x_do_3", "m_do_6", "z_do_6", "x_do_6",
                        "m_do_15", "z_do_15", "x_do_15", "m_do_18", "z_do_18", "x_do_18", "m_sen", "z_sen", "x_sen", "x_do_65"], inplace=True)
                df = df.groupby(["kraj"]).sum()
                df = df.reset_index()
                df["kraj"] = df["kraj"].apply(lambda x: NUTS3_name2abbrev[x])
                df["month"] = month2abbreviations[month]
                df["year"] = int(year)
                df = df.set_index(["month", "year", "kraj"])

                df_non_cummulutive = df.copy()
                #treat cummulative values
                if prev_df is not None:
                    df_non_cummulutive["celkem"] = df["celkem"].values - prev_df["celkem"].values
                    df_non_cummulutive["m_do_65"] = df["m_do_65"].values - prev_df["m_do_65"].values
                    df_non_cummulutive["z_do_65"] = df["z_do_65"].values - prev_df["z_do_65"].values

                return df, df_non_cummulutive
            except:
                print("error:", day, " ", month, " ", year)
```

```python
def prepare_refugees():
    res_df = None
    prev_df = None
    for year, month in zip([2022]*10+[2023]*9, data_num_months):
        df, df_non_cummulative = prepare_month(month, str(year), prev_df=prev_df)
        prev_df = df
        
        if res_df is None:
            res_df = df_non_cummulative
        else:
            res_df = pd.concat([res_df, df_non_cummulative], axis=0)

    return res_df
```

```python
res_df = prepare_refugees()
res_df
```
