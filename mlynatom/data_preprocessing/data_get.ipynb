```python
def fill_min_wage(year, month):
    if year < 2013 or (year == 2013 and month2num[month] < 8):
        return 8000
    elif year < 2015:
        return 8500
    elif year < 2016:
        return 9200
    elif year < 2017:
        return 9900
    elif year < 2018:
        return 11000
    elif year < 2019:
        return 12200
    elif year < 2020:
        return 13350
    elif year < 2021:
        return 14600
    elif year < 2022:
        return 15200
    elif year < 2023:
        return 16200
    else:
        return 17300
    
```

## Refugees - MV

```python
def prepare_month(month: str, year: str, prev_df: Optional[pd.DataFrame]=None) -> pd.DataFrame:
    for day in range(month2days[month], 0, -1):
        for path in [f"https://www.mvcr.cz/soubor/statistika-ukr-{day}-{month}-{year}-xls.aspx", f"https://www.mvcr.cz/soubor/statistika-obce-suk-vek-vse-{day}-{month}-{year}-xls.aspx", f"https://www.mvcr.cz/soubor/statistika-ukr-{day}-{month}-{year}-xlt.aspx"]:
            try:
                df = pd.read_excel(path)
                df.drop(columns=["okres", "obec", "kod_obce", "obec_mc", "m_do_3", "z_do_3", "x_do_3", "m_do_6", "z_do_6", "x_do_6",
                        "m_do_15", "z_do_15", "x_do_15", "m_do_18", "z_do_18", "x_do_18", "m_sen", "z_sen", "x_sen", "x_do_65"], inplace=True)
                df = df.groupby(["kraj"]).sum()
                df = df.reset_index()
                df["kraj"] = df["kraj"].apply(lambda x: NUTS3_name2abbrev[x])
                df["month"] = month2abbreviations[month]
                df["year"] = int(year)
                df = df.set_index(["month", "year", "kraj"])

                df_non_cummulutive = df.copy()
                #treat cummulative values
                if prev_df is not None:
                    df_non_cummulutive["celkem"] = df["celkem"].values - prev_df["celkem"].values
                    df_non_cummulutive["m_do_65"] = df["m_do_65"].values - prev_df["m_do_65"].values
                    df_non_cummulutive["z_do_65"] = df["z_do_65"].values - prev_df["z_do_65"].values

                return df, df_non_cummulutive
            except Exception as e:
                print("error:", e, " ", day, " ", month, " ", year)
```

```python
def prepare_refugees():
    res_df = None
    prev_df = None
    for year, month in zip([2022]*10+[2023]*9, data_num_months):
        df, df_non_cummulative = prepare_month(month, str(year), prev_df=prev_df)
        prev_df = df
        
        if res_df is None:
            res_df = df_non_cummulative
        else:
            res_df = pd.concat([res_df, df_non_cummulative], axis=0)

    return res_df
```

```python
res_df = prepare_refugees()
res_df
```

## Refugees Distributed Lags

```python
df = pd.read_csv("../data/prepared/refugees.csv")
df
```

```python

def compute_distributed_lags(df: pd.DataFrame, kraje: List[str], columns: List[str], window_range: range):
    ret_df = pd.DataFrame(data={})
    for kraj in kraje:
        tmp_df = df[df.kraj == kraj]
        tmp_df.set_index(["month", "year"], inplace=True)
        for column in columns:
            for window in window_range:
                #WARNING min_periods set to 0 (to include the values instead of nans even when smaller numbers are used for sum) -> I think that in this case that make sense -> discuss
                new_col = tmp_df.loc[:, column].rolling(window=window, min_periods=0).sum().to_numpy()
                tmp_df.loc[:, f"{column}_w{window}"] = new_col
                

        tmp_df.reset_index(inplace=True)
        #display(tmp_df)
        ret_df = pd.concat([ret_df, tmp_df], axis=0)

    ret_df.set_index(["month", "year", "kraj"], inplace=True)
    return ret_df
        
        
```

```python
res_df = compute_distributed_lags(df, NUTS3, ["celkem", "m_do_65", "z_do_65"], range(2, 20))
res_df
```
