# Merge DataFrames Together

```python
month2num = {
    "january": 1,
    "february": 2,
    "march": 3,
    "april": 4,
    "may": 5,
    "june": 6,
    "july": 7,
    "august": 8,
    "september": 9,
    "october": 10,
    "november": 11,
    "december": 12,
}
```

```python
import pandas as pd
import os
```

```python
res_df = None
#path = "../data/prepared"
path = "../data/prepared_full/"
for file in os.listdir(path):
    tmp_df = pd.read_csv(os.path.join(path, file), index_col=["month", "year", "kraj"])
    display(tmp_df.head())
    if res_df is None:
        res_df = tmp_df
    else:
        res_df = pd.concat([res_df, tmp_df], axis=1)

display(res_df)
```

```python
def key_fn(x):
    return x.apply(lambda y : month2num[y])
```

```python
res_df.sort_values(by=["kraj"], inplace=True)
res_df.sort_values(by=["month"], inplace=True, key=key_fn, kind="stable")
res_df
```

```python
res_df.sort_values(by=["year"], inplace=True, kind="stable")
```

```python
res_df.to_csv("../data/mlynatom_data_full.csv")
```
