# Merge DataFrames Together

```python
import pandas as pd
import os
```

```python
res_df = None
path = "../data/prepared"
for file in os.listdir("../data/prepared"):
    tmp_df = pd.read_csv(os.path.join(path, file), index_col=["month", "year", "kraj"])
    display(tmp_df.head())
    if res_df is None:
        res_df = tmp_df
    else:
        res_df = pd.concat([res_df, tmp_df], axis=1)

display(res_df)
```

```python
res_df.to_csv("../data/mlynatom_data.csv")
```
